# Análises dos tempos

```python
import pandas as pd
```

## Texto

```python
# Data for text operations (Flutter)
data_text_flutter = {
    "Pergunta": [1, 1, 1, 2, 2, 2],
    "Operação": ["Consulta à Coleção", "Busca por Similaridade", "Execução das Cadeias (chains)",
                 "Consulta à Coleção", "Busca por Similaridade", "Execução das Cadeias (chains)"],
    "Tempo (s)": [0.23, 1.03, 11.33, 0.23, 0.32, 5.35]
}

# Data for text operations (Streamlit)
data_text_streamlit = {
    "Pergunta": [1, 1, 1, 2, 2, 2],
    "Operação": ["Consulta à Coleção", "Busca por Similaridade", "Execução das Cadeias (chains)",
                 "Consulta à Coleção", "Busca por Similaridade", "Execução das Cadeias (chains)"],
    "Tempo (s)": [0.22, 0.28, 4.46, 0.21, 0.25, 5.67]
}
```

```python
df_text_flutter = pd.DataFrame(data_text_flutter)
df_text_flutter
```

```python
df_text_streamlit = pd.DataFrame(data_text_streamlit)
df_text_streamlit
```

```python
# Concatenating data from both applications
combined_data = pd.concat([df_text_flutter, df_text_streamlit])
```

### Calcular o tempo médio por operação (para as perguntas de forma conjunta), combinando os dados de ambas as aplicações (Flutter e Streamlit)

```python
# Calculating the overall mean time per operation, ignoring the question number
mean_time_per_operation_combined = combined_data.groupby('Operação').mean()['Tempo (s)'].reset_index()

mean_time_per_operation_combined
```
